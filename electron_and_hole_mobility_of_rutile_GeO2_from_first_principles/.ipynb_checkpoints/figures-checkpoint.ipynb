## Electron-phonon coupling MEs vs direction for electrons and holes

### Raw data
everything through epw1 is the same as for mob vs temp  
[epw2 input](electron_phonon_coupling/epw2.in)
[epw2 output](electron_phonon_coupling/epw2.out)  
data is processed with [this bash script](electron_phonon_coupling/grab_data.sh)  

Bose-Einstein dist.: n = 1/(exp((hbar x omega)/(kb x T))-1)

### Processed data
```
hbar = 6.582119569e-16 #eV*s
kb = 8.617333262145e-5 #eV/K
kbT = kb*300 #eV
mev2hz = 241799050402.4
fname = 'tmp'
table = {'iq':[],'ibnd':[],'jbnd':[],'imode':[],'enk':[],'omega':[],'g':[],'nq':[]}
with open(fname,'r') as fin:
    while True:
        line = fin.readline()
        if not line:
            break
        else:
            data = line.strip().split()
            
        if data[0] == 'iq':
            iq = int(data[2])
        elif data[0] == 'ibnd':
            trash = fin.readline() #----- line
            data = fin.readline().strip().split()
            while data[0] != '--':
                table['iq'].append(iq)
                table['ibnd'].append(int(data[0]))
                table['jbnd'].append(int(data[1]))
                table['imode'].append(int(data[2]))
                table['enk'].append(float(data[3]))
                table['omega'].append(float(data[5]))
                table['g'].append(float(data[6]))
                omega_hz = float(data[5])*mev2hz
                nq = 1/(np.exp((hbar*omega_hz)/kbT)-1)
                table['nq'].append(nq)
                data = fin.readline().strip().split()
df = pd.DataFrame(table)
df.to_csv('g_data_with_nq.csv',sep=' ',index=False)
```

````python
df = pd.read_csv('electron_phonon_coupling/g_data_with_nq.csv',sep=' ')

a = 2.978360956 #out of plane lattice param (Ang)
bc = 4.515760391 #in plane lattice param (Ang)

#since cell is tetragonal, reciprocal space distances are just inv()
a_inv = 1/a
bc_inv = 1/bc
total_length = a_inv + bc_inv

seg1 = a_inv/total_length
seg2 = bc_inv/total_length

x1 = np.linspace(0,seg1,100)
x2 = np.linspace(seg1,1,99)
xaxis = np.append(x1,x2)
````

````python
def add_label(label,color,size,ax,loc):
    ax.annotate(label,xy=loc,size=size,color=color,xycoords='axes fraction')#,backgroundcolor='w')

def plot_data(ax,xdata,ydata,x_range,y_range,x_label,y_label,formats,legend_label,legend_data):
    if formats == 'default':
        c='k'
        lw=0.95
        ls='-'
        a = 0.5
    else:
        c=formats['c']
        marker=formats['marker']
        try:
            s=formats['s']
        except:
            s=12
        try:
            a=formats['a']
        except:
            a=0.5
   
    ax.scatter(xdata,ydata,c=c,marker=marker,s=s,alpha=a,label=legend_label)

    # style
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
       
    if x_range != 'default':
        ax.set_xlim(x_range)
    if y_range != 'default':
        ax.set_ylim(y_range)

    if legend_label:
        ax.legend()
    if legend_data:
        ax.legend(**legend_data)
````
